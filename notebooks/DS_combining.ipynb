# Balanced Undersampling

```python
import pandas as pd
```

```python
file_dir = "../cleaned_DS"

enron_df = pd.read_csv(f"{file_dir}/enron_cleaned.csv")
print(f"Enron DS total training samples: {enron_df['is_spam'].count().sum()}")
print(enron_df['is_spam'].value_counts())
print()

assassin_df = pd.read_csv(f"{file_dir}/Assassin_cleaned.csv")
print(f"Assassin DS total training samples: {assassin_df['is_spam'].count().sum()}")
print(assassin_df['is_spam'].value_counts())
print()

ceas_df = pd.read_csv(f"{file_dir}/CEAS-08_cleaned.csv")
print(f"CEAS DS total training samples: {ceas_df['is_spam'].count().sum()}")
print(ceas_df['is_spam'].value_counts())
print()
```

```python
ling_df = pd.read_csv(f"{file_dir}/Ling_cleaned.csv")
print(f"Ling DS total training samples: {ling_df['is_spam'].count().sum()}")
print(ling_df['is_spam'].value_counts())
print()

nazario_df = pd.read_csv(f"{file_dir}/Nazario_5_cleaned.csv")
print(f"Nazario DS total training samples: {nazario_df['is_spam'].count().sum()}")
print(nazario_df['is_spam'].value_counts())
print()

nigerian_df = pd.read_csv(f"{file_dir}/Nigerian_5_cleaned.csv")
print(f"Nigerian DS total training samples: {nigerian_df['is_spam'].count().sum()}")
print(nigerian_df['is_spam'].value_counts())
print()
```

```python
# Create a list of all dataframes and concatenate them together
all_dfs = [enron_df, assassin_df, ling_df, nazario_df, nigerian_df, ceas_df]
df_combined = pd.concat(all_dfs)
df_combined.count()
```

```python
# Separate the lists into ham and spam dfs
df_ham = df_combined[df_combined['is_spam'] == 0]
df_spam = df_combined[df_combined['is_spam'] == 1]

# Randomly choose 2.5k samples from each category
ham_sample = df_ham.sample(n=25000, random_state=42)
spam_sample = df_spam.sample(n=25000, random_state=42)

# Combine the samples
df_final = pd.concat([ham_sample, spam_sample])

# Mix the samples and reset the indicies
df_final = df_final.sample(frac=1, random_state=42, ignore_index = True)
print(df_final.head())
print(df_final['is_spam'].value_counts())
```

```python
df_final.to_csv("../cleaned_DS/final.csv", index=False)
```
