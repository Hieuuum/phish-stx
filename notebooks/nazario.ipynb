```python
import pandas as pd

df = pd.read_csv("../datasets/kaggle_phishing_ds2/Nazario_5.csv")
```

```python
import re

def clean_text(s):
    s = str(s).lower()
    s = re.sub(r'http\S+', ' ', s)       # remove URLs
    s = re.sub(r'[^a-z\s]', ' ', s)      # remove punctuation/numbers
    s = re.sub(r"<.*?>", "", s)          # remove HTML tags
    s = re.sub(r'\s+', ' ', s).strip()   # normalize spaces
    return s

df['clean_email_body'] = df['body'].apply(clean_text)
```

```python
df.head()
```

```python
df = df.drop(columns=['sender', 'receiver','date','subject', 'urls','body'])
```

```python
df = df.drop_duplicates(subset=['clean_email_body'])
```
