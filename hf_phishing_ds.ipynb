```python
import re
from collections import Counter

# Optional: clean text inline (recommended)
def clean_text(s):
    s = str(s).lower()
    s = re.sub(r'http\S+', ' ', s)      # remove URLs
    s = re.sub(r'[^a-z\s]', ' ', s)     # remove punctuation/numbers
    s = re.sub(r'\s+', ' ', s).strip()  # normalize spaces
    return s

# Clean if not already cleaned
df['Email Text'] = df['Email Text'].apply(clean_text)

# Separate phishing vs safe emails
phish_texts = " ".join(df.loc[df['Email Type'] == 'Phishing Email', 'Email Text']).split()
safe_texts  = " ".join(df.loc[df['Email Type'] == 'Safe Email', 'Email Text']).split()

```

```python
phish_df = df[df['Email Type'] == 'Phishing Email']
```

```python
phish_df.head()
```

```python

# Combine all phishing emails into one long string of words
phish_words = " ".join(phish_df['Email Text']).split()

# Count how often each word appears
phish_counts = Counter(phish_words)

# Get the 30 most common phishing keywords
top_phish = phish_counts.most_common(30)
print(top_phish)
```
